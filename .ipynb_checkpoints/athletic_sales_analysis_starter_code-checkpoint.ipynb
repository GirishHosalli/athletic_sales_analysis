```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_path = Path('Resources/athletic_sales_2020.csv')
sales_2021_path = Path('Resources/athletic_sales_2021.csv')

sales_2020_df = pd.read_csv(sales_2020_path)
sales_2021_df = pd.read_csv(sales_2021_path)
```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([sales_2020_df, sales_2021_df], join='inner', ignore_index=True)
# combined_sales_df.reset_index
combined_sales_df
```

```python
# Check if any values are null.
combined_sales_df.isnull().values.any()
```

```python
# Check the data type of each column
combined_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_df['invoice_date'] = pd.to_datetime(combined_sales_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_sales_df = combined_sales_df.groupby(['region', 'state', 'city']).agg({'units_sold':'sum'})
grouped_sales_df = grouped_sales_df.sort_values(by='units_sold', ascending=False)
grouped_sales_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
grouped_sales_df.head(5)
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.

pivot_sales_df = pd.pivot_table(combined_sales_df,
                                index=['region', 'state', 'city'],
                                values='units_sold',
                                aggfunc='sum').sort_values(by='units_sold', ascending=False)
# pivot_sales_df = pivot_sales_df.sort_values(by='units_sold', ascending=False)  #descending sort

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_sales_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
pivot_sales_df.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_total_sales_df = combined_sales_df.groupby(['region', 'state', 'city']).agg({'total_sales':'sum'})\
                                          .sort_values(by='total_sales', ascending=False)
# grouped_total_sales_df = grouped_total_sales_df.sort_values(by='total_sales', ascending=False)
grouped_total_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
grouped_total_sales_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_total_sales_df = pd.pivot_table(combined_sales_df,
                                index=['region', 'state', 'city'],
                                values='total_sales',
                                aggfunc='sum')
pivot_total_sales_df = pivot_total_sales_df.sort_values(by='total_sales', ascending=False)  #descending sort

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_total_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
pivot_total_sales_df.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_retailer_sales_df = combined_sales_df.groupby(['retailer', 'region', 'state', 'city']).agg({'total_sales':'sum'})
grouped_retailer_sales_df = grouped_retailer_sales_df.sort_values(by='total_sales', ascending=False)
grouped_retailer_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the top 5 results.
grouped_retailer_sales_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_retailer_sales_df = pd.pivot_table(combined_sales_df,
                                index=['retailer', 'region', 'state', 'city'],
                                values='total_sales',
                                aggfunc='sum')
pivot_retailer_sales_df = pivot_retailer_sales_df.sort_values(by='total_sales', ascending=False)  #descending sort

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_retailer_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
pivot_retailer_sales_df.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
combined_sales_df['product'].unique()
```

```python
# Filter the sales data to get the women's athletic footwear sales data.
# consumables_df = df[df['category'] == top_category]
women_footwear_df = combined_sales_df[combined_sales_df['product'] == "Women's Athletic Footwear"]
women_footwear_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
grouped_women_footwear_df = women_footwear_df.groupby(['retailer', 'region', 'state', 'city']).agg({'units_sold':'sum'})
grouped_women_footwear_df = grouped_women_footwear_df.sort_values(by='units_sold', ascending=False)
grouped_women_footwear_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Rename indexes
grouped_women_footwear_df = grouped_women_footwear_df.rename_axis(index={'retailer': 'Retailer', 'region':'Region', 'state': 'State', 'city': 'City'})

# Show the top 5 results.
grouped_women_footwear_df.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_women_footwear_df = pd.pivot_table(women_footwear_df,
                                         index=['retailer', 'region', 'state', 'city'],
                                         values='units_sold',
                                         aggfunc='sum')
pivot_women_footwear_df = pivot_women_footwear_df.sort_values(by='units_sold', ascending=False)
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_women_footwear_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Rename indexes
pivot_women_footwear_df = pivot_women_footwear_df.rename_axis(index={'retailer': 'Retailer', 'region':'Region', 'state': 'State', 'city': 'City'})

# Show the top 5 results.
pivot_women_footwear_df.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_invoice_date_df = pd.pivot_table(women_footwear_df,
                                       index=['invoice_date'],
                                       values='total_sales',
                                       aggfunc='sum').sort_values(by='total_sales', ascending=False) 
# pivot_invoice_date_df = pivot_invoice_date_df.sort_values(by='total_sales', ascending=False)  # SORT Descending

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_invoice_date_df.rename(columns={"total_sales": "Total Sales"}, inplace=True)
pivot_invoice_date_df.index.name = 'Invoice Date'

# Show the table.
pivot_invoice_date_df.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
pivot_resample_daily_df = pivot_invoice_date_df.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_resample_daily_df = pivot_resample_daily_df.sort_values(by='Total Sales', ascending=False)  #Sort descending
pivot_resample_daily_df.index.name = 'Invoice Date'
pivot_resample_daily_df.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
pivot_resample_weekly_df = pivot_invoice_date_df.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_resample_weekly_df = pivot_resample_weekly_df.sort_values(by='Total Sales', ascending=False)
pivot_resample_weekly_df.index.name = 'Invoice Date'
pivot_resample_weekly_df.head(10)
```
